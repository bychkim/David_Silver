state 안에 또 class를 두어 각 state의 visit 횟수, 각 state에서 action(hit or stick)을 선택한 횟수, state value, action value를 관리한다.

```python
def step(state, action):
    
    """
    Args:
        state (dict) : represent dealer_card and player_total in form of tuple (dealer_card, player_total)
                       state - dealer's first card 1-10 and the player's sum 1-21
        acntion (str) : 'hit' or 'stick'
    
    returns:
        state (object) : next state determined by input action. It could be terminal
        reward (int) : player's win +1, tie 0, lose -1
    """
    
    if action == 'hit':
        card = _draw_card()
        
        # update player's total 
        if card[1] == 'BLACK':
            state.player_total += card[0]
        elif card[1] == 'RED':
            state.player_total -= card[0]
        
        # calculate reward
        if state.player_total > 21 or state.player_total < 1:
            reward = -1
        else:
            reward = 0
            
    elif action == 'stick':
        # dealer's play
        while True:
            card = _draw_card()
            if card[1] == 'BLACK':
                state.dealer_total += card[0]
            elif card[1] == 'RED':
                state.dealer_total -= card[0]
            
            if state.dealer_total >= 17 or state.dealer_total < 1:
                break
            else:
                continue

        # calculate reward 
        if state.dealer_total > 21 or state.dealer_total < 1:
            reward = 1
        elif state.dealer_total < state.player_total:
            reward = 1
        elif state.dealer_total > state.player_total:
            reward = -1
        elif dealer_total == player_total:
            reward = 0

    return state, reward
```

```python
def play(user):
    """playing Easy21
    
    Args:
        user (object) : object which is made by Easy21
    
    """    
    position = 'NONE'
    action = 'NONE'
    
    # initialization of record
    user.win = 0
    user.lose = 0
    user.draw = 0
        
    while(True):
        
        print('-'*30)
        print('Start!')
        # initial state (initialization of state) 
        user.state.player_total = 0
        user.state.dealer_total = 0

        player_start_number = random.randrange(1,11)
        dealer_start_number = random.randrange(1,11)

        user.state.player_total += player_start_number
        user.state.dealer_total += dealer_start_number

        while(True):
            _show_total(user)
            action = input('hit or stick? : ')
            if action == 'quit':
                break
            elif action != 'hit' and action != 'stick':
                print('Invalid Input. Try again.')
                continue
            
            next_reward = step(user.state, action) # tuple

            # arriving at terminal
            if action == 'stick':
                if next_reward[1] == 1:
                    user.win += 1
                elif next_reward[1] == 0:
                    user.draw += 1
                elif next_reward[1] == -1:
                    user.lose += 1
                break
            elif action == 'hit' and next_reward[1] == -1:
                user.lose += 1
                break        
        if action == 'hit' or action == 'stick':
            _show_result(user, next_reward[1])        
        
        if action == 'quit':
            break
```

```python
def _draw_card():
    rand_number = random.randrange(1,11)
    color_number = random.randrange(1,4)
    
    if color_number == 1:
        color = 'RED'
    else:
        color = 'BLACK'
    
    return (rand_number, color)
```

```python
def _show_result(user, recent_reward):
    print('-'*30)
    if recent_reward == 1:
        print('You Win...')
    elif recent_reward == -1:
        print('You Lose...')
    elif recent_reward == 0:
        print('Draw...')
    
    _show_total(user)
    
    print('Win : ', user.win)
    print('Lose : ', user.lose)
    print('Draw :', user.draw)
    print('Total :', user.win + user.lose + user.draw)
    print('Rate : ', round(user.win / (user.win+user.lose+user.draw) * 100, 2), '%')
```

```python
def _show_total(user):
    print('Player\'s total =', user.state.player_total)
    print('Dealer\'s total =', user.state.dealer_total)
    
```
