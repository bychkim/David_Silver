```python
import random
```

```python
def draw_card():
    rand_number = random.randrange(1,11)
    color_number = random.randrange(1,4)
    
    if color_number == 1:
        color = 'RED'
    else:
        color = 'BLACK'
    
    return (rand_number, color)
```

```python
def show_result(win, lose, draw, result, player_total, dealer_total):
    show_total(player_total, dealer_total)
    
    if result == 'WIN':
        print('You Win...')
    elif result == 'LOSE':
        print('You Lose...')
    elif result == "DRAW":
        print('Draw...')
    
    print('Win : ', win)
    print('Lose : ', lose)
    print('Draw :', draw)
    print('Total :', win + lose + draw)
    print('Rate : ', win / (win+lose+draw) * 100, '%')
```
