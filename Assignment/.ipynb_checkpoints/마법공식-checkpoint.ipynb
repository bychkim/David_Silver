```python
import xlrd # excel file read
```

```python
path = r'/Users/bychkim/Documents/파이썬 퀀트 투자/마법공식 데이터.xlsx'

wb = xlrd.open_workbook(path)
```

```python
per_ws = wb.sheet_by_name('PER')
print(per_ws.row_values(0))
print(per_ws.row_values(1))
print(per_ws.row_values(0)[0])
print(per_ws.row_values(0)[1])
```

```python
temp = per_ws.row_values(1)
temp[0], temp[1]
#per_dict[temp[0]] = temp[1]
per_ws.nrows
```

```python
per_dict = {}

for num in range(1, per_ws.nrows):
    temp = per_ws.row_values(num)
    if temp[1] > 0:
        per_dict[temp[0]] = temp[1]
```

```python
per_dict
```

```python
import operator

sorted_per = sorted(per_dict.items(), key=operator.itemgetter(1), reverse=False)
sorted_per
```

```python
per_rank = {}
num = 1

for i in sorted_per:
    #print(i[0])
    per_rank[i[0]] = num
    num = num + 1
```

```python
per_rank
```

```python
roa_ws = wb.sheet_by_name('ROA')

roa_dict = {}
```

```python
for i in range(1, roa_ws.nrows):
    temp = roa_ws.row_values(i)
    if temp[1] is not '':    
        roa_dict[temp[0]] = temp[1]
```

```python
roa_dict
```

```python
sorted_roa = sorted(roa_dict.items(), key=operator.itemgetter(1), reverse=True)
sorted_roa
```

```python
roa_rank = {}
num = 1

for i in sorted_roa:
    #print(i[0])
    roa_rank[i[0]] = num
    num = num + 1
```

```python
roa_rank
```

```python
per_rank['넥스턴'] + roa_rank['넥스턴']
```

```python
total_rank = {}

for firm in roa_rank.keys():
    #print(firm)
    if firm in per_rank.keys():
        #print(roa_rank[firm] + per_rank[firm])
        total_rank[firm] = roa_rank[firm] + per_rank[firm]
```

```python
result = sorted(total_rank.items(), key=operator.itemgetter(1))
```
